```python

df3 = df2
df3['diagnosis'] = df2['diagnosis'].apply(lambda x:1 if x=='M' else 0 )
df3.head()
```

```python
x = df3.drop(['diagnosis'],axis=1)
y = df3['diagnosis']
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression
from sklearn import linear_model
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2)
```

```python
cross_val_score(LogisticRegression(),x_train,y_train)
```

```python
cross_val_score(linear_model.LinearRegression(),x_train,y_train)
```

```python
cross_val_score(DecisionTreeClassifier(),x_train,y_train)
```

```python
cross_val_score(RandomForestClassifier(),x_train,y_train)
```

```python
model = linear_model.LinearRegression()
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
from sklearn.metrics import confusion_matrix 
y_pred = model.predict(x_test)
cm = confusion_matrix(y_pred,y_test)
import seaborn as sn
sn.heatmap(cm,annot = True)
```
